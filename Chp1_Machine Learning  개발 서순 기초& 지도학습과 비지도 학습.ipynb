````python
fish_data = [[l, w] for l, w in zip(length, weight)]
````

배열 자동 배정 및 생성
리스트 내포 - 반복문을 이용한 리스트 생성
다음 조건을 반복 [l,w] for l,w <br>
1차원 배열내 l,w 에 zip 내부 변수 값을 지속 반복 배정<br>
zip 내부의 변수(리스트)의 값이 끝날때까지<br>

````python
fish_target = [1]*35 + [0]*14
````

정답을 먼저 준비. <br>
규칙을 찾을 수 있도록, AI가 학습한 내용이 맞는지 알려주는 <br>
지도학습 방식<br>
도미는 1 빙어는 0으로, 도미가 맞는지 아닌지를 산출<br>

````python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier ()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)
kn.score(fish_data, fish_target)
````

k-최근접 이웃 알고리즘을 이용
a라는 데이터를 훈련 세트와 테스트 세트로 나눈다. <br>
이때, 훈련 세트는 모델을 학습시키기 위한 데이터이고<br> 테스트 세트는 모델이 일반화한 능력을 평가하기 위한 데이터이다.<br>

```
kn = KNeighborsClassifier() #모델 생성
kn.fit(train_input, train_target) #모델 훈련
kn.score(train_input, train_target) #모델 평가, 실행 결과 1.0 출력
```
<br>
위의 코드를 요약하면  <br>훈련 세트로 모델을 훈련하고 평가했고 정확도가 1.0이 출력됨<br>. 모델의 성능이 좋은지 알아보기 위해서는 <br>기존의 데이터(훈련 세트)뿐만 아니라 <br>새로운 데이터(테스트 세트)로 평가할 때에도 정확도가 높아야 한다. <br>그래서 테스트 세트로 모델(훈련 세트로 평가한 모델 kn)을 평가합니다.

`kn.score(test_input, test_target) #모델 평가, 실행 결과 1.0 출력`

실행 결과 1.0이 나왔다는 것은 새로운 데이터(테스트 세트)에서도 <br>정확도가 매우 높다는 것으로 즉 모델 일반화가 잘 되었다 라는 의미

````python
kn.predict([[30, 600]])
````

---

틈새 Tip
텐서(벡터&배열)는 차원이 거듭될수록 부르는 단위-명칭도 다르다.

<p>
3차원까지는 고정.
1dimension = 행 row<br>
2dimension = 열 column<br>
3dimension = 깊이 depth<br>
</p>

<p>
차원별 단위 지칭의 변화 개념<br>
1차원 = 행,렬 존재 x, 단순 "index"로 순서위치 지칭<br>
2차원 = 행,렬 등장, 행[새로]-렬[가로] 순서로 단위 사용

</p>
---


<p>
4차원부터는 목적따라 임의지정<br>
📊 4차원 배열의 목적에 따른 명칭:
이미지 데이터:

batch x channel x height x width

배치(batch): 여러 이미지를 묶어 처리할 때 사용

채널(channel): 각 이미지의 색상 채널(RGB, 그레이스케일 등)

높이(height), 너비(width): 이미지의 크기


---


시계열 데이터:

batch x time x feature x channel

배치(batch): 여러 시계열 데이터를 묶어 처리

시간(time): 시계열 데이터의 시간 단위

특징(feature): 시계열 데이터의 각 특성(예: 주식 데이터의 시가, 종가 등)

채널(channel): 각 시계열의 여러 특성 채널


---


비디오 데이터:

batch x frames x height x width x channel

배치(batch): 여러 비디오를 묶어 처리

프레임(frame): 비디오의 각 시간 단위

높이(height), 너비(width): 비디오 프레임의 크기

채널(channel): 색상 채널(RGB 등)

</p>

````python
import numpy as np
# numpy - 파이썬에 대표적인 배열 라이브러리
# 다수의 Machine Learning 패키지들이 Numpy의 형태로
# 데이터가 제공된다는 것을 상정하고 있음.
# 그만큼 의존적이다 라는 의미

input_arr = np.array(fish_data)
# 학습 배열 = 넘파이 배열 형태로 (원 데이터)
target_arr = np.array(fish_target)
# 정답 배열 =  동일

````

여기서 틈새 Tip<br>
벡터와 배열의 개념 <br>
🔹벡터 (Vector)
수학적 개념

방향과 크기를 가지는 수열

선형대수에서는 공간의 한 점 또는 방향성을 가진 대상

예: [3, 5, -1] → 3차원 벡터

🔹 배열 (Array)
프로그래밍 구조

단순히 데이터를 저장하기 위한 수열 또는 리스트

수학적 의미는 없을 수도 있음

예: [10, 20, 30] → 나이 목록, 가격 리스트 등

🔹 텐서 (Tensor)
벡터, 행렬, 그 이상의 다차원 배열까지 포함하는 상위 개념

0차원(스칼라) ~ n차원까지의 모든 데이터 구조를 포괄

딥러닝에서 핵심 데이터 표현 방식

````python
index = np.arange(49)
# 0~48까지의 수열을 index에 할당 / 특성의 갯수 = 49개
np.random.shuffle(index)
# 순서대로 생성된 수열을 mix
# 여기서, ''_input과 _target 배열을 직접 셔플하지 않는것은
# 각각의 inpurt-target의 값이 서로 쌍을 이루는,매칭이 되어야하는데
# 임의로 섞어버린다면 특성들(input/입력데이터)과 라벨들(target/정답데이터)가
# 매칭이 안되는 엉터리 데이터셋이 되기 때문이다.

train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]
# 셔플한 인덱스를 input,target_arr의 인덱스로 참조
# 각각의 특성,라벨 배열의 값이 매칭이 되도록 해준다.
test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
# 테스트 세트도 동일하게.
````

````python
import matplotlib.pyplot as plt

plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(test_input[:, 0], test_input[:, 1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
````

````python
# 최종 정리

import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# 데이터 정의
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7,
                31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5,
                34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0,
                38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0,
                475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0,
                575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0,
                920.0, 955.0, 925.0, 975.0, 950.0]
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2,
                12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2,
                13.4, 12.2, 19.7, 19.9]

# 데이터 결합
bream_data = [[l, w] for l, w in zip(bream_length, bream_weight)]
smelt_data = [[l, w] for l, w in zip(smelt_length, smelt_weight)]
input_arr = np.array(bream_data + smelt_data)
target_arr = np.array([1]*len(bream_data) + [0]*len(smelt_data))

# 데이터 셔플 및 훈련/테스트 분리
np.random.seed(42)
index = np.arange(len(input_arr))
np.random.shuffle(index)
train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]
print("훈련 특성 갯수",len(train_input))
print("훈련 특성 값",train_input)


test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
print("테스트 특성 갯수",len(test_input))
print("테스트 특성 값",test_input)


# 모델 훈련 및 정확도 측정
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
score = kn.score(test_input, test_target)
print("정확도:", score)
````
